```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

%matplotlib inline
```

```python
class Discrete_Bayesian_Analysis(object):
    def __init__(self,hypos=None,likelihood_function=None):
        self.hypos = hypos
        self.likelihood_function = likelihood_function
    
    def update_hypos(self,prev_dict,data):
        for hypo in self.hypos:
            for word in data:
                try:
                    likelihood = self.likelihood_function(hypo,prev_dict,word)
                    self.hypos[hypo] = self.hypos[hypo] * likelihood
                except:
                    pass
        self.normalize(self.hypos)
        
    def normalize(self,hypos):
        norm_factor = sum(hypos.values())
        for hypo in hypos:
            hypos[hypo] = hypos[hypo]/norm_factor
            
    def plot_hypo_likelihood(self):
        x = [x for x in self.hypos.keys()]
        x.sort()
        
        y = []
        for val in x:
            y.append(self.hypos[val])
        plt.plot(x,y)
```

```python
novel_test = Discrete_Bayesian_Analysis()
```

```python
hypos = {'dracula': 0.5, 'moby_dick': 0.5}


novel_test.hypos = hypos

def word_update(hypo,prev_dict,word):
        return(prev_dict[word][hypo])
    
novel_test.likelihood_function = word_update
```

```python
drac_dict = {}

with open('.\Dracula.txt','r') as f:
    for line in f:
        for word in line.split():
            if(word in drac_dict):
                drac_dict[word] += 1
            else:
                drac_dict[word] = 1 
```

```python
total_words = [x for x in drac_dict.keys()]
```

```python
moby_dict = {}

with open('.\Moby_Dick.txt','r') as f:
    for line in f:
        for word in line.split():
            if(word in moby_dict):
                moby_dict[word] += 1
            else:
                moby_dict[word] = 1
                if(word not in total_words):
                    total_words.append(word)
```

```python
prev_dict = {}
for word in total_words:
    if word in drac_dict:
        drac_count = drac_dict[word]
    else:
        drac_count = 0
    
    if word in moby_dict:
        moby_count = moby_dict[word]
    else:
        moby_count = 0
    
    total_count = drac_count + moby_count
    
    prev_dict[word] = {'dracula':drac_count/total_count,'moby_dick':moby_count/total_count}
```
